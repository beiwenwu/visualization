```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python

# Make a line plot with both of y variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
# Add a legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
# Modify the legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right', 
          frameon = True, #add frame around the legend 
          fontsize = 12, #change font size 
          ncol = 2, #specify number of columns 
          shadow = True )
fig.show()
```

```python
# Move the legend outside of the plot area
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', 
          bbox_to_anchor =(1, 1))
fig.show()
```

```python
# Annotations with the text() function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!")
fig.show()
```

```python
# Modifying text annotations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
        ha='center', ##alignment of text
        color = 'red', ##modify font colour 
        size = 20) ##modify font sizefig.show()

fig.show()
```

```python
# Make a simple, empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# Position text annotations
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# Annotatation with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), 
            arrowprops=dict(facecolor='black', shrink=0.05))
fig.show()
```

```python
# Modify arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', 
            xy=(10, 95), xytext=(20, 94), 
            arrowprops = dict(arrowstyle = 'wedge', 
                              color = 'hotpink'))
```

```python
# Remove tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limit tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Set tick marks at a particular interval (in this case, 5) using MultipleLocator 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotate axis labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right')
```

```python
# Modify axis title fonts
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

font1 = {'family':'serif','color':'indigo'}
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```
